```python
#Content Recommendations based on catgeory
def content_recommendations(username, category, df):
    streamers_within_same_category = df[df['Categories'] == category]
    recommendations = streamers_within_same_category[streamers_within_same_category['Username'] != username]

    return recommendations[['Username', 'Categories']]
```

```python
def category_and_performance_recommendation(username, df, performance_thresholds):
    """
    Function to recommend content creators based on category and performance metrics.

    Parameters:
    - username: The username of the streamer for whom recommendations are needed.
    - df: The DataFrame containing YouTube data.
    - performance_thresholds: A dictionary containing performance thresholds for subscribers, visits, likes, and comments.

    Returns:
    - recommendations: A list of recommended streamers.
    """
    user_row = df[df['Username'] == username]

    # Filter the DataFrame based on category
    same_category_streamers = df[df['Categories'] == user_row['Categories'].values[0]]

    # Filter based on performance metrics
    recommendations = same_category_streamers[
        (same_category_streamers['Suscribers'] > performance_thresholds['subscribers']) &
        (same_category_streamers['Visits'] > performance_thresholds['visits']) &
        (same_category_streamers['Likes'] > performance_thresholds['likes']) &
        (same_category_streamers['Comments'] > performance_thresholds['comments']) &
        (same_category_streamers['Username'] != username)  # Exclude the input streamer
    ]

    return recommendations[['Username', 'Categories', 'Suscribers', 'Visits', 'Likes', 'Comments']]

# Example of using the category and performance-based recommendation system
input_username = 'MrBeast'  # Replace with the actual username

# Set your performance thresholds
thresholds = {
    'subscribers': YouTube_df['Suscribers'].mean(),
    'visits': YouTube_df['Visits'].mean(),
    'likes': YouTube_df['Likes'].mean(),
    'comments': YouTube_df['Comments'].mean(),
}

recommendations = category_and_performance_recommendation(input_username, YouTube_df, thresholds)

print(f"Category and Performance-based Recommendations for {input_username}:")
print(recommendations)
```
